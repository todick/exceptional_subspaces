### Run Subgroup Discovery Method
Possible choices for the method are ```beam-mean```, ```beam-kl```, ```dfs-mean```, ```rsd```, ```syflow```, and ```cart```.

````python
from src.config import Config_RQ2
from src.methods import run_method
config = Config_RQ2()
subgroups, rules = run_method(
    "beam-mean",
    data["data"], 
    data["target"].reshape(-1,1), 
    config=config, 
    n_subgroups=5,
    feature_names=data["feature_names"],
)

plot_subgroups(data["target"], subgroups, rules=rules, which_sgs=[0,1,2,3,4], label_x="Performance", label_y="# configurations", title="", bins=150, stack=False)
````
